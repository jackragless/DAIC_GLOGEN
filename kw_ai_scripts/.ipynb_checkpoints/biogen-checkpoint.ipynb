```python
import pickle
import nltk
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))  
```

```python
with open('/home/jackragless/projects/github/DAIC_glossary_generator/data/wiki_master_dataframe_removed.pkl', 'rb') as f:
    corpus = pickle.load(f)
```

```python
def para_kw_space_remove(curpage):
    curstring = curpage['text']
    for kw in curpage['kw']:
        if curstring.find(kw)!=-1:
            highlight = curstring[curstring.find(kw):curstring.find(kw)+len(kw)].replace(' ', '|/|')
            curstring = str(curstring[:curstring.find(kw)]) + str(highlight) + str(curstring[curstring.find(kw)+len(kw):])
    return curstring
```

```python
def biogen(curpage):
    sents = nltk.sent_tokenize(para_kw_space_remove(curpage))
    kw_arr = curpage['kw']
    final_label_arr = []
    for sent in sents:
        if len(sent)<=512:
            temp_label_arr = []
            sent = sent[:-1]
            for phrase in sent.split():
                count = 0
                if phrase.replace('|/|',' ') in kw_arr:
                    phrase_words = phrase.replace('|/|',' ').split()
                    for word in phrase_words:
                        print(word)
                        if count == 0:
                            temp_label_arr.append('B')
                        else:
                            temp_label_arr.append('I')
                        count+=1
                else:
                    temp_label_arr.append('O')

            final_label_arr.append(temp_label_arr)

        final_sent_arr = []
        for sent in nltk.sent_tokenize(curpage['text']):
            if len(sent)<=512:
                sent = sent[:-1]
                final_sent_arr.append([word for word in sent.split()])

    return final_sent_arr, final_label_arr
```

```python
labels = []
sentences = []
count = 0
for page in corpus[0:300]:
    count += 1
    print(count, end='\r')
    s, l= biogen(page)
    sentences += s
    labels += l
```

```python
for num in range(len(sentences)):
    print(sentences[num],'\n',labels[num],'\n',len(sentences[num]),len(labels[num]),'\n')
```

```python
count = 0
for i in range(len(sentences)):
    if len(sentences[i]) != len(labels[i]):
#         print(test_s[i], len(test_s[i]), test_l[i], len(test_l[i]))
#         print(len)
        count += 1
print(count)
```

```python
final_sent = []
final_label = []
for i in range(len(sentences)):
    temp_sent = []
    temp_label = []
    for j in range(len(sentences[i])):
        if sentences[i][j] not in stop_words:
            temp_sent.append(sentences[i][j])
            temp_label.append(labels[i][j])
    final_sent.append(temp_sent)
    final_label.append(temp_label)
```
